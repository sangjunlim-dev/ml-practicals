```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('possum.csv')
df
```

```python
# Caculates euclidean distance between two points
from sklearn.metrics.pairwise import euclidean_distances

# Calculate pairwise euclidean distance and check the type
type(euclidean_distances(df[['skullw','totlngth']]))
```

```python
# For skullw and totlngth, treat the two measurements for each animal as two-dimensional coordinates (x and y)
distanceMatrix = pd.DataFrame(euclidean_distances(df[['skullw','totlngth']]), index=df['case'], columns=df['case'])

# The pairwise Euclidean distance matrix calculates the similarity or dissimilarity between each animal (case) in the dataset based on their skull width and total length measurements
distanceMatrix
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Split training and test sets
training = df.iloc[2:20]
testing = df.drop(training.index)

# Features: hdlngth, skullw, totlngth, taill
# Target: Trying to predict the 'sex' of the animal
features_col = ['hdlngth','skullw','totlngth','taill']
target_col = 'sex'
```

```python
# Initialize classifier with k=3
classifier = KNeighborsClassifier(n_neighbors=3)

# Train the training data with input features and target variable
classifier.fit(training[features_col], training[target_col])
```

```python
# Predict sex for each animal in the testing set with trained KNN classifier
pd.Series((classifier.predict(testing[features_col])))
```

```python
# Add the predictions to the test set
testing['predictionUsingKnn'] = classifier.predict(testing[features_col])
```

```python
# Confusion_matrix
from sklearn.metrics import f1_score, accuracy_score, precision_score, recall_score, confusion_matrix

# Rows are actual values, Columns are predicted values
pd.DataFrame(confusion_matrix(testing['sex'], testing['predictionUsingKnn']))
```

```python
# Calculate the accuracy of the KNN model
accuracy_score(testing['sex'], testing['predictionUsingKnn'])
```
