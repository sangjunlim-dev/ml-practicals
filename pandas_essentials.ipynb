### Apply Functions

````python
df7 = df.copy()

# Apply functions to each column (operate on rows - axis=0)
df7.apply(np.max)

df7[df7['CCA3'].apply(lambda code: code.startswith('C'))].head()

# Replace values in a column with map and dictionary
# Similar to replace function, however this method will change them to NaN if value isn't present in the dictionary
df7['TF'] = df7['CCA3'].apply(lambda code: 'Yes' if code.startswith('C') else 'No')
d = {'Yes': True, 'No': False}

df7['TF'] = df7['TF'].map(d)
````

### Grouping
```df.groupby(by=GROUPING_COLUMN)[COLUMN_TO_SHOW].function()```

````python
df8 = df.copy()

# The grouping column becomes a new index
df8.groupby(['Continent']).describe()

df8.groupby(['Continent']).agg({'2022 Population': ['mean']}).sort_values(by=('2022 Population', 'mean'), ascending=False)
````

### Summary Tables

````python
df9 = df.copy()

df9['2022 Population'].mean()

df9['Avg Population'] = (df9['2022 Population'] > 34632250).astype('int')

pd.crosstab(df9['Continent'], df9['Avg Population'], margins=True)
````
