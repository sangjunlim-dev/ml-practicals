## Linear Regression

#### Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

#### Load Sample Dataset

```python
X, y = datasets.load_diabetes(return_X_y=True)
```

#### Select a Feature

```python
X = X[:, np.newaxis, 2]
X
```

#### Data split: Training and Test sets

```python
split = int(len(X) * 0.75)

X_train = X[:split]
X_test = X[split:]

y_train = y[:split]
y_test = y[split:]
```

#### Plot Training Set

```python
plt.scatter(X_train, y_train)
plt.title("Scatter Plot of Training Data")
plt.xlabel("X")
plt.ylabel("y")

plt.xticks(())
plt.yticks(())

plt.show()
```

#### Fit the Linear Regression Model

```python
model = LinearRegression()
model.fit(X_train, y_train)

intercept = model.intercept_
coefficient = model.coef_

print(f"Intercept: {intercept}")
print(f"Coefficient: {coefficient}")
```

#### Predictions and Evaluation of the Model

```python
predictions = model.predict(X_test)

mse = mean_squared_error(y_test, predictions)
print("Mean Squared Error:", round(mse, 3))

r2 = r2_score(y_test, predictions)
print("R-sqaured:", round(r2, 3))
```

#### Predictions vs. Acutal Values

```python
plt.scatter(X_test, y_test, color='blue', label='Actual Values')
plt.plot(X_test, predictions, color='red', linewidth=1, label='Regression Line')
plt.title("Linear Regression Fit with Test Data")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()

plt.xticks(())
plt.yticks(())

plt.show()
```
